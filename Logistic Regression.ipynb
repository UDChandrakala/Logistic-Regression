Importing Libraries

```python
import numpy as np
import pandas as pd
```

Importing Dataset

```python
df = pd.read_csv('Social_Network_Ads.csv')
```

```python
df
```

```python
x = df.iloc[:,:-1].values
```

```python
x
```

```python
y=df.iloc[:,-1].values
```

```python
y
```

Split Dataset

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
len(x_train)
```

```python
len(x_test)
```

scale the 2 nd cloumn since it ranges from 0 to 1 lakh

Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
x_train = sc.fit_transform(x_train) #transforming values in between from 0 to 1
x_test = sc.fit_transform(x_test)
```

```python
x_train
```

Model

```python
from sklearn.linear_model import LogisticRegression
```

```python
classifier = LogisticRegression(random_state = 0)
```

```python
classifier.fit(x_train,y_train)
```

```python
#32	18000	0
```

Prediction

```python
print(classifier.predict(sc.transform([[32,15000]]))[0])
```

```python
print(classifier.predict(sc.transform([[20,86000]]))[0])
```

```python
y_pred = classifier.predict(x_test)
```

```python
y_pred
```

```python
print(np.concatenate((y_pred.reshape(len(y_pred),1),y_test.reshape(len(y_test),1)),1))
```

Confusion Matrix and Accuracy

```python
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
cm = confusion_matrix(y_test,y_pred)
```

```python
cm  # here 49 and 20 are the correct predicted values                                             
```

```python
accuracy_score(y_test,y_pred)
```
